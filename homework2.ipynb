You are helping an IT department upgrade its corporate network. The network is represented by a tree $T = (V, E)$.

You have a budget to upgrade routers at $k$ nodes of the tree. If you upgrade the equipment at _both_ endpoints of a link between adjacent nodes $u$ and $v$, the bandwidth of the link improves, and, as a result, user productivity increases by $p_{uv}$ points.

Your goal is to identify $k$ nodes to upgrade so as to maximize the total gain in user productivity. In other words, you need to design an algorithm that finds a subset $S \subseteq V$ of $k$ vertices that maximizes
$$
\sum_{\substack{(u,v) \in E \\ u,v \in S}} p_{uv}.
$$

You should implement your algorithm in a function _FindBestUpgrade_. The function receives a full binary tree with $n = 2^h - 1$ vertices labeled $0, \dots, 2^h - 2$.
The tree is rooted at $0$. Each non-leaf vertex $i$ has two children, $2i+1$ and $2i+2$. The values of the parameters $p_{ij}$ are stored in an array $p$:
- $p[2i]$ stores the value of $p_{(i,\,2i+1)}$;
- $p[2i+1]$ stores the value of $p_{(i,\,2i+2)}$.


Your function should output the value
$$
\sum_{\substack{(u,v) \in E \\ u,v \in S^*}} p_{uv}
$$
of the optimal solution $S^*$.
