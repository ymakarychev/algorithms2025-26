Consider the following problem. You are given a sequence $a_0, \dots, a_{n-1}$.

The goal is to find an integer $k \ge 0$ and a subsequence $c_0, \dots, c_{2k-1}$ of $a_0, \dots, a_{n-1}$ that maximizes the following expression:
$$
\sum_{i=0}^{k-1} c_{2i} c_{2i+1}.
$$

Design and implement a DP algorithm for the problem. Your program should return the value of the optimal sequence
$$
\sum_{i=0}^{k-1} c_{2i} c_{2i+1}
$$
(if $k = 0$, the value of the sum is 0). The running time should be $O(n^2)$.

---

### Example 1

$a = (a_0, \dots, a_9) = (4, 3, -4, 3, 2, 5, -3, -5, 1, -4)$.

The optimal solution is $c = (4, 3, 3, 5, -5, -4)$:

$$
\color{red}{\mathbf{4}}\quad
\color{red}{\mathbf{3}}\quad
-4\quad
\color{red}{\mathbf{3}}\quad
2\quad
\color{red}{\mathbf{5}}\quad
-3\quad
\color{red}{\mathbf{-5}}\quad
1\quad
\color{red}{\mathbf{-4}}
$$

Its value is
$$
4 \cdot 3 + 3 \cdot 5 + (-5) \cdot (-4) = 47.
$$
